```python
import pandas as pd
import re
from numpy import arange

replicate = 5
f=open("TEST" + str(replicate) + ".txt", "r")
lines = f.readlines()
f.close()

_time = []
_uwb_l = []
_uwb_r = []

# data parse : list in time // cop // imu data :  parse acc.x,y,z : gyro.x,y,z : mag.x,y,z
for line in lines :
    time = line.split(',')
    _time.append(time[0])
    _uwb_l.append(time[11])
    _uwb_r.append(time[13])


with open("Test" + str(replicate) + ".csv", "w") as out_file:
    for i in range(len(_uwb_r)):
        out_string = ""
        out_string += str(_time[i])
        out_string += "," + str(_uwb_r[i])
        out_string += "," + str(_uwb_l[i])
        out_string += "\n"
        out_file.write(out_string)
        
    
```

```python
import matplotlib.pyplot as plt
# data received and graph


plt.plot(_time,_uwb_r,'r')
plt.plot(_time,_uwb_l,'b')

plt.xlabel('time')
plt.ylabel('uwb')
plt.show()
```
